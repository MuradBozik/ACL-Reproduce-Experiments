```python
import numpy as np
import pandas as pd
import re
```

```python
csv_folder = "scenarios/CaDiCaL/slices-BMC08/"
csv_name = "configs_old_Copy.csv"

df1 = pd.read_csv(csv_folder + csv_name, header=None, sep="'(.*?)'", engine="python")
df1.head()
```

```python
df1.drop(columns=np.arange(len(df1.columns), step=2), inplace=True)
df1.head()
```

```python
with open(csv_folder + csv_name, 'r') as csvfile:
    one_line_text = csvfile.readline()

regr =" -(.*?) " 

col_names = re.findall(regr,one_line_text)
df1.columns = col_names
df1.head()
```

```python
df2 = pd.read_csv(csv_folder+csv_name, header=None, sep="p(.*?) ", engine="python")
df2[1].head()
```

```python
df1.insert(0, "parameter", df2[1])
df1.head()
```

```python
def manipulate_df(df, target):
    
    cols = list(df.columns[1:])
    
    cols.remove(target)
    
    target_values = list(df1.loc[df1.parameter.apply(lambda x:x.split('-')[0]) == target.upper(), target])
        
    with open(csv_folder+csv_name, 'r') as csvfile:
        regr = target + " '(.*?)'" 

        lines = csvfile.readlines()
        new_lines = []
        for i in range(len(lines)):
            result = re.search(regr,lines[i]).span()
            
            for ix, value in enumerate(target_values):
                line_id,line_conf= lines[i].split(',')
                new_line_id = line_id + '_' + target.upper() + '-'+str(ix) + ','
                
                new_lines.append(new_line_id+line_conf.replace(lines[i][result[0]:result[1]],target+ f" '{value}'"))
                
        
    with open(csv_folder+"configs_new.csv", "w") as f:
        f.write("#Config ID, configuration call string\n")
        
        for line in new_lines:
            if line.startswith('p'+target.upper()):
                pass
            else:
                f.write(line)
    return df
```

```python
target ="keepglue"
df3 = manipulate_df(df1, target)
df3
```
