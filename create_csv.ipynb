```python
import numpy as np
import pandas as pd
import re
```

```python
csv_path = "scenarios/CPLEX/slices-cls/configs_Copy-1.csv"

df1 = pd.read_csv(csv_path, header=None, sep="'(.*?)'", engine="python")
df1.head()
```

```python
df1.drop(columns=np.arange(len(df1.columns), step=2), inplace=True)
df1.head()
```

```python
with open(csv_path, 'r') as csvfile:
    one_line_text = csvfile.readline()

regr =" -(.*?) " 

col_names = re.findall(regr,one_line_text)
df1.columns = col_names
df1.head()
```

```python
df2 = pd.read_csv(csv_path, header=None, sep="p(.*?)-", engine="python")
df2[1].head()
```

```python
df1.insert(0, "parameter", df2[1])
df1.head()
```

```python
def manipulate_df(df, target):
    
    cols = list(df.columns[1:])
    
    cols.remove(target)
    
    target_values = list(df.loc[df.parameter == target.upper(), target])
    
    for col in cols:
        df.loc[df.parameter == col.upper(), target] = list(target_values)[:len(df.loc[df.parameter == col.upper()])]
        
    with open(csv_path, 'r') as csvfile:
        regr = target + " '(.*?)'" 

        lines = csvfile.readlines()
        for i in range(len(lines)):
            result = re.search(regr,lines[i]).span()
            lines[i] = lines[i].replace(lines[i][result[0]:result[1]],target+ f" '{df.iloc[i,df.columns.get_loc(target)]}'")
    
        
    with open("scenarios/CPLEX/slices-cls/configs_new.csv", "w") as f:
        f.write("#Config ID, configuration call string\n")
        
        for line in lines:
            if line.startswith('p'+target.upper()):
                pass
            else:
                f.write(line)
    return df
```

```python
target ="mip_limits_submipnodelim"
df3 = manipulate_df(df1, target)
df3
```
