```python
df4 = pd.read_csv("scenarios/EAX/slices-tsp-rue-1000-3000/responses/response-npop.csv")
df4
```

```python
from matplotlib import pyplot as plt
import numpy as np

#some example data
x= df4['#Parameter value']
y = df4[' median of Mean running time']
#some confidence interval
lb = df4[' 95% lower bound on mean running time']
ub = df4[" 95% upper bound on mean running time"]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_ylim([0.00025,0.0005])
ax.fill_between(x, lb, ub, color='b', alpha=.1)
ax.set_title("Parameter slices reponses for npop")
ax.set_ylabel("CPU Seconds")
ax.set_xlabel("NPOP Values")
plt.savefig("response_npop_scaled.png")
plt.show()
```
